```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

path = "/content/data/network-intrusion-dataset/Monday-WorkingHours.pcap_ISCX.csv"
df = pd.read_csv(path)

print(df.shape)
df.head()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.strip()
```

```python
df["Label"].value_counts()
```

```python
# load datasets
benign = pd.read_csv("/content/data/network-intrusion-dataset/Monday-WorkingHours.pcap_ISCX.csv")
attack = pd.read_csv("/content/data/network-intrusion-dataset/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv")

# clean columns
benign.columns = benign.columns.str.strip()
attack.columns = attack.columns.str.strip()
```

```python
# label
benign["y"] = 0
attack["y"] = 1

# combine
df = pd.concat([benign, attack], ignore_index=True)

features = [
    "Flow Duration",
    "Total Fwd Packets",
    "Total Backward Packets",
    "Flow Bytes/s",
    "Flow Packets/s",
    "Fwd Packet Length Mean",
    "Bwd Packet Length Mean"
]

# numeric + clean invalid
X = df[features].apply(pd.to_numeric, errors="coerce")
X = X.replace([np.inf, -np.inf], 0).fillna(0)
y = df["y"]
```

```python
# train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# model
model = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
# evaluation
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=["BENIGN", "PORTSCAN"]))
```

```python
import joblib

joblib.dump(model, "portscan_rf_model.pkl")
joblib.dump(features, "features.pkl")
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

features = [
    "Flow Duration",
    "Total Fwd Packets",
    "Total Backward Packets",
    "Flow Bytes/s",
    "Flow Packets/s",
    "Fwd Packet Length Mean",
    "Bwd Packet Length Mean",
    "Active Mean",
    "Idle Mean"
]

X = df[features].apply(pd.to_numeric, errors="coerce").values
y = df["Label"].apply(lambda x: 0 if x == "BENIGN" else 1).values

# ðŸ”§ critical fix
X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train = X_scaled[y == 0]
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import EarlyStopping

input_dim = X_train.shape[1]

inp = Input(shape=(input_dim,))
x = Dense(32, activation="relu")(inp)
x = Dense(16, activation="relu")(x)
x = Dense(32, activation="relu")(x)
out = Dense(input_dim)(x)

autoencoder = Model(inp, out)
autoencoder.compile(optimizer="adam", loss="mse")

autoencoder.fit(
    X_train, X_train,
    epochs=30,
    batch_size=256,
    validation_split=0.1,
    callbacks=[EarlyStopping(patience=3)],
    verbose=1
)
```

```python
recon = autoencoder.predict(X_scaled)
recon_error = np.mean((X_scaled - recon) ** 2, axis=1)

threshold = np.percentile(recon_error[y == 0], 95)

pred = (recon_error > threshold).astype(int)
```

```python
pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Threshold:", threshold)
print(classification_report(y, pred, target_names=["BENIGN", "ATTACK"]))
print(confusion_matrix(y, pred))
```
